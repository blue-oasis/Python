```python
class setADT :
  def __init__(self) :
    self.L = set([])

  def size(self) :
    print("원소의 갯수는 : ", len(self.L))
    return len(self.L)

  def contains(self, a) :
    if a in self.L : 
      print(a , "is Ture")
      return True
    else : 
      print(a , "is False")
      return False

  def insert(self, a) : 
    if a in self.L : 
      print("set에 이미", a ," 가 존재합니다.")
    else : 
      print("set에", a , "넣었습니다")
      self.L.add(a)

  def update(self, *args) : 
    self.L.update(args)
    print(self.L)

  def delete(self, a) :
    if a in self.L : 
      print("set에서 ", a , "를 꺼냈습니다")
      self.L.remove(a)
      print(self.L)
    else : 
      print("set에", a , "가 존재하지 않습니다.")

  def equalus(self, a) :
    if self.L == a : 
      print(a, "와 같습니다.")
    else :
      print(a, "와 다릅니다.")

  def union(self, a) :
    b = self.L.union(a)
    self.L = b
    print(self.L)

  def intersect(self, a) :
    b = self.L.intersection(a)
    self.L = b
    print(self.L)

  def differense(self, a) :
    b = self.L.difference(a)
    self.L = b
    print(self.L)

  def display(self) : 
     print("set : " ,self.L)

b = {1, 2, 3 ,11, 14}

a = setADT()
a.insert(0)
a.insert(2)
a.update(3, 4, 5)
a.size()
a.delete(1)
a.equalus({1, 2, 3})
a.union(b)
a.intersect(b)
a.differense({11, 14})
a.display()

```

```python
class Stack :
  def __init__(self):
    self.L = []
  
  def isEmpty(self):
    if len(self.L) == 0 :
      print("스택이 비어있습니다.")
      return True
    else :
      print("스택에 원소가 들어있습니다.")
      return False

  def push(self, e):
    self.L.append(e)
    print("원소", e, "추가했습니다.")
    print(self.L)

  def update(self, e) : 
    self.L.extend(e)
    print(self.L)
  
  def pop(self) :
    num2 = self.L.pop()
    print(num2 ,"를 꺼내고 비웠습니다.")
    print(self.L)
    return num2

  def peek(self):
    if len(self.L) != 0 : 
      num = len(self.L) - 1
      num2 = self.L[num]
      print(num2 ,"를 꺼냈습니다.")
      return num2

  def size(self) :
    print("원소의 갯수는 : ", len(self.L))
    return len(self.L)

  def clear(self) :
    self.L.clear()
    print("스택을 비웠습니다.")


s = Stack()
s.push(1)
s.push(2)
s.pop()
s.peek()
s.size()
s.clear()


  
```
