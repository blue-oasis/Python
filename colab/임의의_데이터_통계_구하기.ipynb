```python
import random
import numpy as np


def randomNum(num1, num2, num3) :
  '''랜덤범위 숫자 num1~num2/ 출력 숫자 갯수 num3
      int 리스트 return'''
  L = []

  for i in range(num1) :
    r_n = random.randint(num2, num3)
    L.append(r_n)

  return L

def LtoS(num1 = 5, num2 = 1, num3 = 10) :
  '''랜덤범위 숫자 num1~num2/ 출력 숫자 갯수 num3
     랜덤숫자 숫자 리스트 string 변환'''
  L = randomNum(num1, num2, num3)
  return ','.join(map(str, L))

num = [] #숫자 갯수, 범위 리스트

def numInput():
  '''숫자 갯수, 범위 입력 함수''' 
  print("초기값 : 1~10 범위의 숫자 5개 생성")
  inputL = []
  while True:
    num = list(map(int, input("숫자 갯수(n1), 범위(n2~n3)를 입력해 주세요 (0 입력시 종료,입력 없을 시 기본값 유지) : ").split()))
    
    if(len(num)==0):
      L = LtoS()
      inputL.append(L)
      continue
    
    if (num[0] == 0): 
      break
    
    if(len(num) == 1):
      L = LtoS(num[0])
    elif(len(num) == 2):
      L = LtoS(num[0], num[1])
    else :
      L = LtoS(num[0], num[1], num[2])
    inputL.append(L)

  return inputL

name = ""

def fileInput(L): 
  '''파일 이름 지정, 데이터 입력'''
  global name
  name = input("파일 이름을 정해주세요")
  with open(name+".csv", "w+", encoding='utf-8-sig') as f :
    #f.write("숫자 쓰기"+ "\n")
    for i in L:
      f.write(i + "\n")

def clc(name) :
  '''파일 한 줄씩 읽어들여서 계산 가능하게 변환''' 
  clcL = []
  with open(name+".csv", "a+", encoding='utf-8-sig') as f :
    cL = f.readlines()
    for i in cL :
      S = i.replace('\n', '')
      S = S.replace(',' , ' ')
      intList = list(map(int, S.split())) 
      clcL.append(intList)
    
   #함수 다만들고 호출해야디~
    return clcL

def fun1(L):
  '''평균, 분산, 표준편차 2중리스트'''
  L1 = []
  for i in L : 
    AVG = sum(i)/len(i)
    VAR = np.var(i)
    STD = np.std(i)
    L1.append([AVG, VAR, STD, "\n"])
  return L1

def numTOstring(L):
  '''fun1 결과물 str변환'''
  S = str(L)
  S = S.replace('[', '')
  S = S.replace(']', '')
  S = S.replace('\'', '')

  return S

def w(L):
  with open(name+".csv", "w+", encoding='utf-8-sig') as f :
    f.write(L)

L1 = numInput()        
fileInput(L1)
L2 = fun1(clc(name))
st = numTOstring(L2)
w(st)
```
